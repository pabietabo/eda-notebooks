## Data Visualization Basics (Pandas and Seaborn)

```python
top20 = df_small.sort_values("Bevlkerung", ascending=False).head(20)

top20.set_index("Stadtteil")["Bevlkerung"].plot(
    kind="bar",
    figsize=(12,6),
    title="Top 20 Neighbourhoods in Hamburg by Population"
)
plt.ylabel("Population")
plt.show()
```

```python
# Reshape data so we can use hue for comparison
df_long = top20.melt(
    id_vars="Stadtteil",
    value_vars=["65-Jhrigeundltere", "Unter18-Jhrige"],
    var_name="Age Group",
    value_name="Bevlkerung"
)

# Plot
plt.figure(figsize=(14,6))
sns.barplot(
    x="Stadtteil", 
    y="Bevlkerung", 
    hue="Age Group", 
    data=df_long
)

plt.title("Population Under 18 and Over 65 by Neighbourhood")
plt.xticks(rotation=90)  # rotate labels so they dont overlap
plt.ylabel("Population")
plt.xlabel("Neighbourhood")
plt.show()
```

```python
df_small['Durchschnitt-lichesEinkommenjeSteuerpflichtigeninEUR(2010)'].hist(bins=30, edgecolor='black')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
df_long = top20.melt(
    id_vars="Stadtteil",
    value_vars=["65-Jhrigeundltere", "Unter18-Jhrige"],
    var_name="Age Group",
    value_name="Population"
)

plt.figure(figsize=(8,6))
sns.boxplot(x="Age Group", y="Population", data=df_long)
plt.title("Distribution of Age Groups Across Top 20 Neighbourhoods")
plt.show()
```

## Basic Feature Engineering (Augmentation)

```python
df['IsWealthy'] = df['Durchschnitt-lichesEinkommenjeSteuerpflichtigeninEUR(2010)'].apply(lambda x: 1 if x >= 70000 else 0)
df[['Stadtteil', 'Durchschnitt-lichesEinkommenjeSteuerpflichtigeninEUR(2010)', 'IsWealthy']].head(10)
```

```python
df['IsImpoverished'] = df['Durchschnitt-lichesEinkommenjeSteuerpflichtigeninEUR(2010)'].apply(lambda x: 1 if x <= 16000 else 0)
df[['Stadtteil', 'Durchschnitt-lichesEinkommenjeSteuerpflichtigeninEUR(2010)', 'IsImpoverished']].head(10)
```

```python
df['AgedRatio'] = df['65-Jhrigeundltere'] / df['Bevlkerung'] 
df[['Stadtteil', 'Bevlkerung', '65-Jhrigeundltere', 'AgedRatio']].head()
```

```python
# Show correlation between numerical features
corr = df_small.corr(numeric_only=True)

# Heatmap
plt.figure(figsize=(20,12))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Use pairplot
sns.pairplot(df_small[["Bevlkerung", "PrivatePKW(Jan2016)", "65-Jhrigeundltere", ]])
```

```python
# Bin car density
max_density = df_small["PKW-Dichte(Jan2016)"].max()
bins = [0, 200, 400, 600, 800, max_density]

# Make sure sorted & unique
bins = sorted(set(bins))

df_small["CarDensityCat"] = pd.cut(
    df_small["PKW-Dichte(Jan2016)"],
    bins=bins,
    labels=[f"{bins[i]}{bins[i+1]}" for i in range(len(bins)-1)],
    include_lowest=True
)

# Bin unemployment
max_unemp = df_small["Arbeitslosenanteilin%(Dez2015)"].max()
bins = [0, 5, 10, 15, 20, max_unemp]

# Ensure bins are strictly increasing
bins = sorted(set(bins))

df_small["UnemploymentCat"] = pd.cut(
    df_small["Arbeitslosenanteilin%(Dez2015)"],
    bins=bins,
    labels=[f"{bins[i]}{bins[i+1]}" for i in range(len(bins)-1)],
    include_lowest=True
)

# Countplot: how unemployment categories distribute across car density
plt.figure(figsize=(10,6))
sns.countplot(
    x="CarDensityCat",
    hue="UnemploymentCat",
    data=df_small
)
plt.title("Neighbourhood Counts by Car Density and Unemployment Category")
plt.xlabel("Car Density (per 1,000 Residents)")
plt.ylabel("Number of Neighbourhoods")
plt.show()
```
